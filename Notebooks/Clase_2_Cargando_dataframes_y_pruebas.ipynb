##**Exploratory Data Analysis (EDA)**

¿Qué es el preprocesamiento de datos?

El preprocesamiento de datos es el proceso de convertir los datos de su forma original en una forma ordenada y organizada para que estén listos para el análisis y libres de problemas.

El preprocesamiento de datos es importante en el proceso de ciencia de datos e incluye muchas etapas, como la importación de datos, la limpieza de datos, el procesamiento de cadenas y la extracción de texto, además de la gestión de datos faltantes y la gestión de fechas y horas.

### **Preprocesamiento**

Explorando datos

```python
df = pd.read_csv('https://raw.githubusercontent.com/mn-ariel/AUX_DS_ML_2_2023/main/datasets/Pokemon.csv')

df.head()
```

### **Los datos:**

_**#**_: ID para cada pokemon


**Name**: Nombre de cada pokemon

**Type 1**: Cada pokemon tiene un tipo, esto determina la debilidad/resistencia a los ataques

**Type 2**: Algunos pokemon son de tipo dual y tienen 2

**Total**: suma de todas las estadísticas que vienen después de esto, una guía general de qué tan fuerte es un pokemon

**HP**: puntos de golpe, o salud, define cuánto daño puede soportar un pokemon antes de desmayarse

**Attack**: el modificador base para ataques normales (por ejemplo, Scratch, Punch)

**Defense** la resistencia base al daño contra ataques normales

**Sp Atk**: ataque especial, el modificador base para ataques especiales (por ejemplo, explosión de fuego, haz de burbujas)

**Sp Def**: la resistencia base al daño contra ataques especiales

**Speed**: determina qué pokemon ataca primero en cada ronda

**Generation** determina a que genereacion pertenecen

**Legendary** determina si un pokemon es o no es legendario

Vamos a modificar el nombre de las columnas (cabezaras) para tener un mayor facilidad al momento de trabajar con ellas

### 3.  **Comprobar datos**

Para verificar el conjunto de datos en busca de NA, estoy usando isna(), que devuelve un objeto booleano del mismo tamaño que indica si los valores son NA y luego estoy calculando la cantidad de valores verdaderos para cada variable.

```python
all_data_NA = df.isna().sum()

pd.concat([all_data_NA], axis=1, sort = False, keys = ['Todos NA'])
```

```python
pokemon_type_1 = pd.DataFrame(df.groupby("type_1").number.nunique().reset_index(name="count"))

pokemon_type_2 = pd.DataFrame(df.groupby("type_2").number.nunique().reset_index(name="count"))

pd.concat([pokemon_type_1, pokemon_type_2], axis=1, sort = False, keys = ['', ''])
```

### 4.  **Visualizar datos**

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

Matplotlib es una biblioteca popular de visualización de datos para Python que permite crear gráficos, diagramas y otros tipos de visualizaciones personalizables.

Seaborn es una biblioteca de visualización de datos para Python que se construye sobre Matplotlib y proporciona una sintaxis más simplificada y fácil de usar para crear gráficos estadísticos.

```python
fig = plt.figure(figsize=(12, 8), dpi=100)

plt.subplot(2,1,1)
ax = plt.bar(pokemon_type_1["type_1"], pokemon_type_1["count"])
plt.xticks(rotation = 30)
plt.title("Type 1")

plt.subplot(2,1,2)
ax = plt.bar(pokemon_type_2["type_2"], pokemon_type_2["count"])
plt.title("Type 2")

plt.xticks(rotation = 30)

fig.suptitle('Pokemon by type: Type 1 and 2', fontsize=16)

plt.subplots_adjust(hspace=0.4)

plt.show()
plt.clf()
```

```python
df["types_combos"] = df["type_1"] + " " + df["type_2"].fillna('')
df["types_combos"] = df.types_combos.str.strip()

df.head(5)
```

```python
pokemon_combo_type = df.groupby("types_combos").number.nunique().reset_index(name='count')
pokemon_combo_type_25 = pokemon_combo_type.sort_values(["count"], ascending = False).head(25)
```

```python
plt.figure(figsize=(12, 10), dpi=100)
ax = plt.barh(pokemon_combo_type_25["types_combos"], pokemon_combo_type_25["count"])
plt.gca().invert_yaxis()
plt.title("Pokemon by type: including combinations  (Top 25)")
plt.show()
plt.clf()
```

```python
pokemon_combo_type_generations = df.groupby(["types_combos", "generation"]).number.nunique().reset_index(name='count')
pokemon_combo_type_generations_1 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 1].sort_values(["count"], ascending = False).head(25)
pokemon_combo_type_generations_2 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 2].sort_values(["count"], ascending = False).head(25)
pokemon_combo_type_generations_3 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 3].sort_values(["count"], ascending = False).head(25)
pokemon_combo_type_generations_4 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 4].sort_values(["count"], ascending = False).head(25)
pokemon_combo_type_generations_5 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 5].sort_values(["count"], ascending = False).head(25)
pokemon_combo_type_generations_6 = pokemon_combo_type_generations[pokemon_combo_type_generations["generation"] == 6].sort_values(["count"], ascending = False).head(25)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 3, figsize=(18, 18), dpi=150)


plt.subplot(2,3,1)
ax = plt.barh(pokemon_combo_type_generations_1["types_combos"], pokemon_combo_type_generations_1["count"])
plt.gca().invert_yaxis()
plt.title("Generation 1")

plt.subplot(2,3,2)
ax = plt.barh(pokemon_combo_type_generations_2["types_combos"], pokemon_combo_type_generations_2["count"])
plt.gca().invert_yaxis()
plt.title("Generation 2")

plt.subplot(2,3,3)
ax = plt.barh(pokemon_combo_type_generations_3["types_combos"], pokemon_combo_type_generations_3["count"])
plt.gca().invert_yaxis()
plt.title("Generation 3")

plt.subplot(2,3,4)
ax = plt.barh(pokemon_combo_type_generations_4["types_combos"], pokemon_combo_type_generations_4["count"])
plt.gca().invert_yaxis()
plt.title("Generation 4")

plt.subplot(2,3,5)
ax = plt.barh(pokemon_combo_type_generations_5["types_combos"], pokemon_combo_type_generations_5["count"])
plt.gca().invert_yaxis()
plt.title("Generation 5")


plt.subplot(2,3,6)
ax = plt.barh(pokemon_combo_type_generations_6["types_combos"], pokemon_combo_type_generations_6["count"])
plt.gca().invert_yaxis()
plt.title("Generation 6")


fig.suptitle('Pokemon types by generation: Including combo types (Top 25)', fontsize=20)

plt.subplots_adjust(
                    wspace= 0.3, 
                    hspace= 0.1,
                    top = 0.93,
                    bottom = 0)


plt.show()
plt.clf()
```

```python
legendary = df.groupby("legendary").number.nunique()
explode = [0, 0.2]



plt.figure(figsize=(16, 6), dpi=100)
plt.pie(legendary, autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("How many pokemons are Legendary?")
plt.show()
plt.clf()
```

```python
legendary_generation = df.groupby(["legendary", "generation"]).number.nunique().reset_index(name='count')
legendary_gen1 = legendary_generation[legendary_generation["generation"] == 1].sort_values(["count"], ascending = False).head(25)
legendary_gen2 = legendary_generation[legendary_generation["generation"] == 2].sort_values(["count"], ascending = False).head(25)
legendary_gen3 = legendary_generation[legendary_generation["generation"] == 3].sort_values(["count"], ascending = False).head(25)
legendary_gen4 = legendary_generation[legendary_generation["generation"] == 4].sort_values(["count"], ascending = False).head(25)
legendary_gen5 = legendary_generation[legendary_generation["generation"] == 5].sort_values(["count"], ascending = False).head(25)
legendary_gen6 = legendary_generation[legendary_generation["generation"] == 6].sort_values(["count"], ascending = False).head(25)
```

```python
explode = [0, 0.2]

fig, (ax1, ax2) = plt.subplots(2, 3, figsize=(14, 14), dpi=120)

plt.subplot(2,3,1)
plt.pie(legendary_gen1["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 1")

plt.subplot(2,3,2)
plt.pie(legendary_gen2["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 2")

plt.subplot(2,3,3)
plt.pie(legendary_gen3["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 3")

plt.subplot(2,3,4)
plt.pie(legendary_gen5["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 4")

plt.subplot(2,3,5)
plt.pie(legendary_gen5["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 5")

plt.subplot(2,3,6)
plt.pie(legendary_gen6["count"], autopct='%1.1f%%', labels = ["Non-legendary", "Legendary"], pctdistance=0.6, colors = ["#d6c372","#bec7c7"], explode = explode, startangle = 60)
plt.title("Generation 6")


fig.suptitle('Legendary pokemons across generations', fontsize=20)


plt.subplots_adjust(
                    wspace= 0, 
                    hspace= 0,
                    top = 1,
                    bottom = 0.2)

plt.show()
plt.clf()
```

```python
legendary_by_type = df[df.legendary == True].groupby("types_combos").number.nunique().reset_index(name='count')
legendary_by_type_sorted = legendary_by_type.sort_values(by = "count", ascending=False)
legendary_by_type_sorted
```

```python
legendary_by_type_sorted_top6 = legendary_by_type_sorted[:6].copy()

new_row = pd.DataFrame(data = {"types_combos": ["Others"], "count": [legendary_by_type_sorted["count"][6:].sum()]})

legendary_by_type_sorted_top6 = pd.concat([legendary_by_type_sorted_top6, new_row])
legendary_by_type_sorted_top6
```

```python
plt.figure(figsize=(16, 6), dpi=100)
plt.pie(legendary_by_type_sorted_top6["count"], labels = ["Psychic none", "Water none", "Normal none", "Dragon Psychic", "Electric Flying", "Fire Flying", "Others (only 1 per type combo)"], autopct='%1.1f%%', pctdistance=0.8, startangle = 60)
plt.title("Which types are the Legendary Pokemon?")
plt.show()
plt.clf()
```
