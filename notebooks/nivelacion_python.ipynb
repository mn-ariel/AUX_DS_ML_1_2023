```python
def sumador(numero):
    auxiliar = numero + 5
    print(auxiliar)
```

```python
def restador(numero):
    aux = numero - 1
    print(aux)
```

```python
restador(15)
```

```python
def duplicador_fruitful(numero):
  resultado = numero * 2
  return(resultado)
```

```python
def duplicador_void(numero):
  resultado = numero * 2
  print(resultado)
```

```python
var_1 = duplicador_fruitful(9) 
print(var_1)
print(var_1 + 5)

```

```python
duplicador_void(9)
```

```python
def promedio_dos_notas_void(nota_1, nota_2):
  promedio = (nota_1 + nota_2)/2
  print(promedio)
```

```python
promedio_dos_notas_void(60,100)
```

Digamos que tiene puntos extra por participar

```python
promedio_dos_notas_void(60,100) + 5
```

```python
def promedio_dos_notas_fruitful(nota_1, nota_2):
  promedio = (nota_1 + nota_2)/2
  return(promedio, nota_1, nota_2)
```

```python
resultados = promedio_dos_notas_fruitful(60, 100)
print(resultados)
```

```python
for i in resultados:
  print(i + 5)
```

## Condicionales

### If

```python
edad_pablo = 18
edad_mateo = 15
```

```python
if(edad_pablo >= 18):
  print("Puede sacar licencia de conducir")
else:
  print("Debe esperar un tiempo m√°s")
```

```python
if(edad_mateo > 18):
  print("Puede sacar licencia de conducir")
else:
  print("Debe esperar un tiempo m√°s")
```

```python
def edad_licencia(edad):
  if edad >= 18:
    print("Puede sacar licencia de conducir")
  else:
    print("Debe esperar un tiempo m√°s")
```

```python
edad_licencia(edad_pablo)
edad_licencia(edad_mateo)
```

### Una condicional m√°s

```python
def operadora(opcion):
  '''
  opcion 1: contactarse con asistencia
  opcion 2: contactarse con servicio tecnico
  opcion 3: contactarse con el gerente
  '''
  if (opcion == 1):
    print("Se te contactara con asistencia")
  
  elif (opcion == 2):
    print("Se te contactara con servicio tecnico")
  
  elif (opcion == 3):
    print("Se te contactara con gerente")
  
  else:
    print("Hasta luego")
```

```python
operadora(1)
```
